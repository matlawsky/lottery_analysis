```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
draws = pd.read_csv('cleaned_data.csv')
```

```python
draws.columns.unique()
```

```python
sns.countplot(x = 'day_of_the_week', data = draws)
plt.show()
```

```python
nr1 = pd.Series(draws['nr1'])
nr2 = pd.Series(draws['nr2'])
nr3 = pd.Series(draws['nr3'])
nr4 = pd.Series(draws['nr4'])
nr5 = pd.Series(draws['nr5'])
nr6 = pd.Series(draws['nr6'])
numbers = pd.DataFrame(pd.concat([nr1,nr2,nr3,nr4,nr5,nr6]), columns=["number"])
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.countplot(x = numbers['number'], data = numbers)
plt.show()
```

On the diagram is portrayed amount of time sigle number appeared as a wining number, we can observe that some numbers appeared more frequently than others, when analyzing the reasons we should explain how the lottery exactly works:
